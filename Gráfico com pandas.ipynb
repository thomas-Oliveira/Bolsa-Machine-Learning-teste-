# Gráficos usando a biblioteca pandas

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

### df.hist ( )

Mostra a distribuição dos dados

```python
df2 = pd.DataFrame({'a': np.random.randn(1000) + 1, 
                     'b': np.random.randn(1000),
                     'c': np.random.randn(1000) - 1}, 
                      columns=['a', 'b', 'c'])
df2.head(10)
```

```python
#alpha muda a intensidade das cores
plt.figure();
df2.plot.hist(alpha=0.5);
```

### df.plot ( ) 

Usando assim ele gera um gráfico de linhas.

```python
velocidade = [0.1, 17.5, 40, 48, 52, 69, 88]
anos_de_vida = [2, 8, 70, 1.5, 25, 12, 28]
animais = ['caracol', 'porco', 'elefante','coelho', 'girafa', 'coiote', 'cavalo']

df = pd.DataFrame({'velocidade': velocidade, 'anos de vida': anos_de_vida}, index=animais)
df
```

```python
df.plot(title = 'Gráfico 1');
```

### plot.barh( )

```python
df.plot.barh();
```

```python
#stacked é para empilhar as barras
df.plot.bar(rot = 0, stacked = True);
```

### plot.bar( )

```python
df.plot.bar();
```

```python
df.plot.bar(rot = 0, stacked = True);
```

#### Dividindo as informações em vários gráficos

```python
df.plot.bar(rot = 0, subplots = True);
```

### plot.scatter ( )

Cria um gráfico de dispersão.

```python
df = pd.DataFrame(np.random.rand(100, 3), columns=['a', 'b', 'c'])
df.head(10)
```

```python
#sendo c os valores para colorir os pontos
df.plot.scatter(x='a',
                y='b',
                c='c',
                colormap='viridis');
```

### plot.pie ( )

Cria um gráfico de pizza.

```python
df = pd.DataFrame({'massa': [0.330, 4.87 , 5.97],
                   'raio': [2439.7, 6051.8, 6378.1]},
                   index=['Mercúrio', 'Venus', 'Terra'])
df
```

```python
#figsize determina o tamnho da imagem
df.plot.pie(y='massa', figsize=(5, 5));
```

```python
plot = df.plot.pie(subplots=True, figsize=(12, 6))
```
